```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Updated sample data with more subjects
data = {
    "Subject": ["Maths", "Maths", "Maths", "Maths", "Maths", 
                "Science", "Science", "Science", "Science", "Science",
                "Social Studies", "Social Studies", "Social Studies", "Social Studies", "Social Studies",
                "English", "English", "English", "English", "English"],
    "Hours Studied": [10, 15, 20, 25, 30, 8, 9, 5, 1, 3, 10, 8, 6, 4, 2, 7, 5, 3, 9, 11]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x="Subject", y="Hours Studied", data=df)
plt.title("Student Performance: Hours Studied by Subject")
plt.xlabel("Subject")
plt.ylabel("Number of Hours Studied")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset
np.random.seed(10)
data = pd.DataFrame({
    'value': np.concatenate([np.random.normal(0, 1, 100), np.random.normal(10, 1, 10)])
})

# Calculate IQR for box plot based outlier detection
Q1 = data['value'].quantile(0.25)
Q3 = data['value'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data[(data['value'] < lower_bound) | (data['value'] > upper_bound)]
print(f"Outliers based on Box Plot criteria: \n{outliers}")

# Create a box plot and violin plot to visualize the data and outliers
plt.figure(figsize=(12, 6))

# Box plot
plt.subplot(1, 2, 1)
sns.boxplot(x=data['value'])
plt.title('Box Plot')

# Violin plot
plt.subplot(1, 2, 2)
sns.violinplot(x=data['value'])
plt.title('Violin Plot')

plt.show()
```
